## Question Answering using LLM with Prompt Engineering

```python
combinations = [
    {
        "name": "Combo 1: Zero-Shot (Precise)",
        "instruction": "Answer the medical question as a professional clinician. Be concise and factual.",
        "params": {"temperature": 0.0, "top_p": 0.95}
    },
    {
        "name": "Combo 2: Persona-Based (Dermatologist/Specialist)",
        "instruction": "Act as a specialized Senior Medical Consultant. Provide detailed clinical symptoms and standard treatments.",
        "params": {"temperature": 0.2, "top_p": 0.90}
    },
    {
        "name": "Combo 3: Chain of Thought (Reasoning)",
        "instruction": "Think step-by-step. First, analyze the symptoms, then identify the likely condition, and finally outline the standard treatment protocol.",
        "params": {"temperature": 0.1, "top_p": 0.85}
    },
    {
        "name": "Combo 4: Constraint-Based (Bullet Points)",
        "instruction": "Provide the answer using only a structured bulleted list. Do not use full paragraphs.",
        "params": {"temperature": 0.0, "top_p": 0.95}
    },
    {
        "name": "Combo 5: Comprehensive (Exploratory)",
        "instruction": "Explain the condition, common causes, and both medical and surgical interventions if applicable.",
        "params": {"temperature": 0.5, "top_p": 0.99}
    }
]
```

```python
def response_engineered(query, instruction, params):
    # Mistral Instruction Format
    prompt = f"<s>[INST] <<SYS>>\\n{instruction}\\n<</SYS>>\\n\\n{query} [/INST]"

    # Generate output using the LLM object defined in your previous section
    model_output = llm(
        prompt=prompt,
        max_tokens=512,
        temperature=params['temperature'],
        top_p=params['top_p'],
        echo=False
    )

    return model_output['choices'][0]['text'].strip()
```

## Data Preparation for RAG

### Loading the Data

```python
pdf_path = "/content/drive/My Drive/medical_diagnosis_manual.pdf"

loader = PyMuPDFLoader(pdf_path)
data = loader.load()

print(f"Successfully loaded {len(data)} pages from Google Drive.")
```

### Data Overview

#### Checking the first 5 pages

```python
for i in range(5):
    print(f"--- Page {i+1} ---")
    print(data[i].page_content[:500]) # Print first 500 characters of each page
    print("\n")
```

#### Checking the number of pages

```python
print(f"Total number of pages in the manual: {len(data)}")
```

### Data Chunking

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    add_start_index=True
)

chunks = text_splitter.split_documents(data)

print(f"Total number of chunks created: {len(chunks)}")
```

### Embedding

```python
mbedding_model = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

### Vector Database

```python
vector_db = Chroma.from_documents(
    documents=chunks,
    embedding=mbedding_model,
    persist_directory="./chroma_db"
)

print("Vector database created and persisted.")
```

### Retriever

```python
retriever = vector_db.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}
)
```

### System and User Prompt Template

```python
qna_system_message = """
You are a highly accurate Medical AI Assistant.
Your task is to provide clinical information based ONLY on the provided Merck Manual context.
If the information is not present in the context, clearly state that the manual does not contain this specific information.
"""

qna_user_message_template = """
### Context from Merck Manual:
{context}

### Question:
{question}

### Instructions:
Provide a detailed answer based on the context above. Use professional medical terminology.

### Answer:
"""
```

### Response Function

```python
def generate_rag_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input,k=k)
    context_list = [d.page_content for d in relevant_document_chunks]

    # Combine document chunks into a single context
    context_for_query = ". ".join(context_list)

    user_message = qna_user_message_template.replace('{context}', context_for_query)
    user_message = user_message.replace('{question}', user_input)

    prompt = qna_system_message + '\n' + user_message

    # Generate the response
    try:
        response = llm(
                  prompt=prompt,
                  max_tokens=max_tokens,
                  temperature=temperature,
                  top_p=top_p,
                  top_k=top_k
                  )

        # Extract and print the model's response
        response = response['choices'][0]['text'].strip()
    except Exception as e:
        response = f'Sorry, I encountered the following error: \n {e}'

    return response
```

## Output Evaluation

Let us now use the LLM-as-a-judge method to check the quality of the RAG system on two parameters - retrieval and generation. We illustrate this evaluation based on the answeres generated to the question from the previous section.

- We are using the same Mistral model for evaluation, so basically here the llm is rating itself on how well he has performed in the task.

```python
groundedness_rater_system_message  = """
You are an expert medical evaluator. Your task is to determine if the provided Answer is GROUNDED in the given Context.
- 'Grounded' means the answer uses ONLY information found in the context.
- 'Not Grounded' means the answer includes information NOT found in the context or contradicts the context.
Respond with a single word: 'Grounded' or 'Not Grounded', followed by a one-sentence explanation.
"""
```

```python
relevance_rater_system_message ="""
You are a medical quality assurance specialist. Your task is to determine if the provided Answer is RELEVANT to the User's Question.
- 'Relevant' means the answer directly addresses the symptoms, treatments, or protocols asked for.
- 'Irrelevant' means the answer is off-topic or fails to answer the specific question asked.
Respond with a single word: 'Relevant' or 'Irrelevant', followed by a one-sentence explanation.
"""
```

```python
user_message_template = """
### Context:
{context}

### Question:
{question}

### Answer:
{answer}

### Evaluation:
"""
```

```python
def generate_ground_relevance_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input,k=3)
    context_list = [d.page_content for d in relevant_document_chunks]
    context_for_query = ". ".join(context_list)

    # Combine user_prompt and system_message to create the prompt
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""

    response = llm(
            prompt=prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    answer =  response["choices"][0]["text"]

    # Combine user_prompt and system_message to create the prompt
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Combine user_prompt and system_message to create the prompt
    relevance_prompt = f"""[INST]{relevance_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    response_1 = llm(
            prompt=groundedness_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    response_2 = llm(
            prompt=relevance_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    return response_1['choices'][0]['text'],response_2['choices'][0]['text']
```
